```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# ---------------------------------------------
# PART 1: Single ODE dy/dt = -ky using odeint
# ---------------------------------------------

# Parameters
k = 0.3
y0 = 5  # Initial condition

# Define the ODE: dy/dt = -k*y
def single_ode(y, t):
    return -k * y

# Time points for solving
t = np.linspace(0, 10, 100)  # Time range from 0 to 10

# Solve the ODE using odeint
y_solution = odeint(single_ode, y0, t)

# Plot the solution
plt.figure(figsize=(8, 5))
plt.plot(t, y_solution[:, 0], label='y(t)', color='blue')
plt.title("Solution of dy/dt = -ky")
plt.xlabel("Time t")
plt.ylabel("y(t)")
plt.legend()
plt.grid()
plt.show()
```

## Romeo and Juliet, a Dynamical Love Affair in Python

We consider a romatic affair between  Romeo and Juliet, where R(t) is Romeo’s love (or hate if negative) for Juliet at time t and J(t) is Juliet’s love for Romeo.

**dR/dT=aR+bJ**

**dJ/dT=cR+dJ**

where a  and  b  specify  Romeo’s  love,  and  c  and  d  specify  Juliet’s love. Parameter  a  describes  the  extent  to  which  Romeo  is  determined  by  his  own  feelings,  and  b  is  the  extent  to  which  he  is determined  by  Juliet’s  feelings. Likewise,  d  describes  the  extent  to  which  Juliet  is  determined  by  her  own  feelings,  and  b  is  the  extent  to  which  she  is determined  by  Romeo’s  feelings.

Parameters a and d can be thought of as "behavioral inertia", and b and c as "influence function". The  resulting  dynamics  are  two-dimensional,   governed   by   the   initial   conditions   and   the   four   parameters, which may be positive or negative.


To implement this model in Python, we first import the necessary libraries. We will use [odeint](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.odeint.html) to solve the differential equations.

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# ---------------------------------------------
# Coupled System of ODEs for Romeo and Juliet
# ---------------------------------------------

# Parameters
a, b, c, e = 0, 0.4, -0.3, 0  # Given parameters
R0, J0 = 3.14, -0.5           # Initial conditions: R(0) and J(0)

# Define the system of ODEs
def coupled_odes(X, t):
    R, J = X
    dR_dt = a * R + b * J  # dR/dt
    dJ_dt = c * R + e * J  # dJ/dt
    return [dR_dt, dJ_dt]

# Time span for the solution
t = np.linspace(0, 50, 500)  # Time points from t=0 to t=50

# Solve the system of ODEs using odeint
initial_conditions = [R0, J0]
solution = odeint(coupled_odes, initial_conditions, t)

# Extract R(t) and J(t)
R = solution[:, 0]
J = solution[:, 1]

# ---------------------------------------------
# Plot the Solutions
# ---------------------------------------------

plt.figure(figsize=(10, 6))
plt.plot(t, R, label="R(t): Romeo's love", color='red')
plt.plot(t, J, label="J(t): Juliet's love", color='green')
plt.title("Coupled ODEs for Romeo and Juliet's Love Dynamics")
plt.xlabel("Time t")
plt.ylabel("Love/Hate Level")
plt.legend()
plt.grid()
plt.show()
```

The plot illustrates the oscillatory dynamics of Romeo's and Juliet's emotions, resembling a harmonic oscillator. Romeo's love R(t) (red curve) and Juliet's love J(t) (green curve) exhibit periodic cycles of love and hate, but they are out of phase. When Romeo's love peaks, Juliet's feelings cross the neutral point, and vice versa. This phase shift highlights their emotional interdependence but lack of synchronization.
